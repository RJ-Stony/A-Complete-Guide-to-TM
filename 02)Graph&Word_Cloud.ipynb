## Word Frequency Graph - Which Words Are Most Used?

```python
import nltk
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

from nltk.corpus import gutenberg
file_names = gutenberg.fileids()
# Read the file title.

print(file_names)
```

```python
doc_alice = gutenberg.open('carroll-alice.txt').read()
print('# Num of characters used:', len(doc_alice))     # Number of characters used
print('# Text sample:')
print(doc_alice[:500])     # Print only the first 500 characters.
```

```python
from nltk.tokenize import word_tokenize
tokens_alice = word_tokenize(doc_alice)     # Tokenization Execution

print('# Num of tokens used:', len(tokens_alice))
print('# Token sample:')
print(tokens_alice[:20])
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

# Running Stemming for All Tokens
stem_tokens_alice = [stemmer.stem(token) for token in tokens_alice]

print('# Num of tokens after stemming:', len(stem_tokens_alice))
print('# Token sample:')
print(stem_tokens_alice[:20])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Running Stemming for All Tokens
lem_tokens_alice = [lemmatizer.lemmatize(token) for token in tokens_alice]

print('# Num of tokens after lemmatization:', len(lem_tokens_alice))
print('# Token sample:')
print(lem_tokens_alice[:20])
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer("[\w']{3,}")

reg_tokens_alice = tokenizer.tokenize(doc_alice.lower())
print('# Num of tokens with RegexpTokenizer:', len(reg_tokens_alice))
print('# Token sample:')
print(reg_tokens_alice[:20])
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords     # Words that are not normally analyzed
english_stops = set(stopwords.words('english'))     # Convert to a non-repeating set.

# Create a list with only words excluding stopwords.
result_alice = [word for word in reg_tokens_alice if word not in english_stops]

print('# Num of tokens after stopword elimination:', len(result_alice))
print('# Token sample:')
print(result_alice[:20])
```

```python
alice_word_count = dict()
for word in result_alice:
  alice_word_count[word] = alice_word_count.get(word, 0) + 1

print('# Num of used words:', len(alice_word_count))

sorted_word_count = sorted(alice_word_count, key=alice_word_count.get, reverse=True)

print('# Top 20 high frequency words:')
for key in sorted_word_count[:20]:     # Print the top 20 words
  print(f'{repr(key)}: {alice_word_count[key]}', end=', ')
```

```python
my_tag_set = ['NN', 'VB', 'VBD', 'JJ']
my_words = [word for word, tag in nltk.pos_tag(result_alice) if tag in my_tag_set]

alice_word_count = dict()
for word in my_words:
  alice_word_count[word] = alice_word_count.get(word, 0) + 1

print('# Num of used words:', len(alice_word_count))

sorted_word_count = sorted(alice_word_count, key=alice_word_count.get, reverse=True)

print("# Top 20 high frequency words:")
for key in sorted_word_count[:20]:     # Print the top 20 words
  print(f'{repr(key)}: {alice_word_count[key]}', end=', ')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Create a list by getting the frequencies for a sorted list of words.
w = [alice_word_count[key] for key in sorted_word_count]

plt.plot(w)
plt.show()
```

```python
n = sorted_word_count[:20][::-1]     # Extract the top 20 words and sort them in reverse order.
w = [alice_word_count[key] for key in n]     # Frequency for 20 words
plt.barh(range(len(n)), w, tick_label=n)     # Horizontal bar graph
plt.show()
```

## Graph and Word Cloud for Korean documents

```python
!pip install konlpy
```

```python
from konlpy.corpus import kolaw
const_doc = kolaw.open('constitution.txt').read()

print(type(const_doc))  # Check the type of imported data
print(len(const_doc))
print(const_doc[:600])
```

```python
from konlpy.tag import Okt
t = Okt()
tokens_const = t.morphs(const_doc)  # Tokenize by Morpheme

print('# 토큰의 수:', len(tokens_const))
print('# 앞 100개의 토큰')
print(tokens_const[:100])
```

```python
tokens_const = t.nouns(const_doc)   # Extract only nouns after tokenizing by morpheme unit
print('# 토큰의 수:', len(tokens_const))
print('# 앞 100개의 토큰')
print(tokens_const[:100])
```

```python
tokens_const = [token for token in tokens_const if len(token) > 1]
print('# 토큰의 수:', len(tokens_const))
print('# 앞 100개의 토큰')
print(tokens_const[:100])
```

```python
from matplotlib import font_manager, rc

# In case of Mac, specify font_name as follows
# font_name = 'AppleGothic'
rc('font', family='NanumBarunGothic')

const_cnt = {}
for word in tokens_const:
  const_cnt[word] = const_cnt.get(word, 0) + 1

def word_graph(cnt, max_words=10):

  sorted_w = sorted(cnt.items(), key=lambda kv: kv[1])
  print(sorted_w[-max_words:])
  n, w = zip(*sorted_w[-max_words:])

  plt.barh(range(len(n)), w, tick_label=n)
  plt.savefig('bar.png')
  plt.show()

word_graph(const_cnt, max_words=20)
```

```python
font_path = 'malgun.ttf'
wordcloud = WordCloud(font_path=font_path).generate(const_doc)

plt.axis("off")
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```

```python
wordcloud = WordCloud(
    font_path = font_path,
    max_font_size = 100,
    width = 800,
    height = 400,
    background_color='white',
    max_words=50)

# Create a Word Cloud from the result of morphological analysis rather than the original text
wordcloud.generate_from_frequencies(const_cnt)

wordcloud.to_file("const.png")

plt.axis("off")
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
